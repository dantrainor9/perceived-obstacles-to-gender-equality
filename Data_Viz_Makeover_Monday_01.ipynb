```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
```

```python
filename = '/content/drive/MyDrive/CodingDojo Data Science Bootcamp/03 Week 3 Data Visualization/_Viz5_Feb_2021_data - Perceived Obstacles to Gender Equality Focus 2030  Women Deliver .xlsx'
df = pd.read_excel(filename)
df
```

```python
response_dictionary = {'Because women and men are not equally represented in politics' : 'Political representation', 'Because of religion and culture that do not treat women and men as equals': 'Culture and religious views', 'Because boys and girls are treated differently growing up': 'Differential treatment as children', 'Because women and men have different employment opportunities': 'Different employment opportunities', 'Because unpaid care, domestic work, and parental responsibilities are not shared equally between women and men': 'Unequal distribution of domestic labor', 'Because women and men cannot exercise the same level of control over their bodies (for example access to contraception, delaying childbirth)': 'Body autonomy', 'Because crises (for example conflict, natural disasters, pandemics) have unequal effects on women and men' : 'Unequal consequences of large scale crises', 'Because men tend to be physically stronger than women' : 'Differences in physical strength', 'Because girls do not receive the same access to education as boys' : 'Unequal access to education', 'None of these' : 'None of these', "Don't know": "Don't know"}
df['Response']=df['Response'].apply(lambda x: response_dictionary[x])
```

```python
responses = df.groupby(['Response', 'Country']).mean().reset_index()
responses = responses.sort_values(by='% of respondents')

sns.barplot(data=responses, x='Response', y='% of respondents', hue='Country')
plt.xticks(rotation=90)
plt.legend(loc=(1.02,0));
```

```python
totalfilter = df.loc(['Gender'])=='Total'


gender = df.groupby(['Response', 'Gender']).mean().reset_index()
gender = gender.sort_values(by='% of respondents')

sns.barplot(data=gender, x='Response', y='% of respondents', hue='Gender')
plt.xticks(rotation=90)
plt.legend(loc=(1.02,0));
```
